```python
date['date']=pd.to_datetime(date['date'])
```

```python
# extracting year
date['year']=date['date'].dt.year
```

```python
# extracting month
date['month']=date['date'].dt.month
```

```python
# extracting month_name
date['month_name']=date['date'].dt.month_name()
```

```python
# extracting date
date['date_day']=date['date'].dt.day
```

```python
# EXTRACTING DAY OF WEEK
date['day_of_week']=date['date'].dt.dayofweek
```

```python
# extracting name of week
date['day_name']=date['date'].dt.day_name()
```

```python
# is weekend
date['weekend']=np.where(date['day_of_week'].isin([5,6]),1,0)
```

```python
date.sample(5)
```
