```python
# figure level function
# catplot
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# strip plot
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# figure level function catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='scatter')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
```

```python
# strip using catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='strip')
```

```python
# jitter
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='strip',jitter=False)
```

```python
# boxplot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box',col='smoker')
```

```python
# Box plot +KDE plot
# Violin plot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# split ->True
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='violin',split=True)
```

```python
# barplot
sns.barplot(data=tips,x='day',y='total_bill')
```

```python
# hue
sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
#error bar
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',errorbar=None,estimator='std')
```
